```python
#Import Libraries

import pandas as pd
import numpy as np
```

### Import Data

```python

df = pd.read_csv('ecom-data.csv',encoding= 'unicode_escape')
```

### Remove Transactions with Zero Quantity

```python
df = df.loc[df['Quantity'] > 0]
```

```python
df['CustomerID'].describe()
```

### Remove Records with Missing Customer ID

```python
df['CustomerID'].isna().sum()
```

```python
df = df.dropna(subset=['CustomerID'])
```

### Post Cleaning Data

```python
df.shape
```

```python
df.head()
```

### Transform Data into Matrix

```python
customer_item_matrix = df.pivot_table(
    index='CustomerID', 
    columns='StockCode', 
    values='Quantity',
    aggfunc='sum'
)
```

```python
customer_item_matrix.head()
```

### Replace values with 0 & 1

```python
customer_item_matrix = customer_item_matrix.applymap(lambda x: 1 if x > 0 else 0)
```

```python
customer_item_matrix.head()
```

# Item-based Collaborative Filtering

### Import Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
item_item_sim_matrix = pd.DataFrame(cosine_similarity(customer_item_matrix.T))
```

```python
item_item_sim_matrix.columns = customer_item_matrix.T.index

item_item_sim_matrix['StockCode'] = customer_item_matrix.T.index
item_item_sim_matrix = item_item_sim_matrix.set_index('StockCode')
```

```python
item_item_sim_matrix.head()
```

### Top 10 Products Similar to Item Code 23168

```python
top_10_similar_items = list(
    item_item_sim_matrix\
        .loc['23168']\
        .sort_values(ascending=False)\
        .iloc[:10]\
    .index
)
```

```python
top_10_similar_items
```

```python
df.loc[
    df['StockCode'].isin(top_10_similar_items), 
    ['StockCode', 'Description']
].drop_duplicates().set_index('StockCode').loc[top_10_similar_items]
```

### Top 10 Products Similar to Item Code 10124A

```python
top_10_similar_items_10124A = list(
    item_item_sim_matrix\
        .loc['10124A']\
        .sort_values(ascending=False)\
        .iloc[:10]\
    .index
)
```
