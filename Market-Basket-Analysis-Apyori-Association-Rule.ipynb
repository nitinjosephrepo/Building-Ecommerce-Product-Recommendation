```python
# Remove the Column Names
df = pd.read_csv('store_data.csv',header=None)
```

```python
# Look at the shape
print(df.shape)
```

```python
#Converting pandas dataframe into a list of lists
records = []
for i in range(0, len(df)):
    records.append([str(df.values[i,j]) for j in range(0, len(df.columns))])
```

```python
#Import apyori library
from apyori import apriori
```

```python
# Build the Apyori model
associations = apriori(records, min_support=(7 * 5) / len(df), min_confidence=0.2,
                       min_lift=3, min_length=2)
```

```python
#Converting the associations to lists
associations = list(associations)
```

```python
#Print out the number of rules
print(len(associations))
```

```python
#Lets check the first association rule
print(associations[0])
```

```python
# Print All Association rules
for item in associations:
    # first index of the inner list
    # Contains base item and add item
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])
    #second index of the inner list
    print("Support: " + str(item[1]))
    #third index of the list located at 0th
    #of the third index of the inner list
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
